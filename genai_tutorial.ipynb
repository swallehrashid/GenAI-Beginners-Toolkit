This notebook demonstrates how to use Python to generate text using a pre-trained AI model.

```python
# Install the necessary library (We will use Hugging Face transformers for free access)
%pip install transformers torch
```

```python
from transformers import pipeline

# Initialize the text generation pipeline
# We are using GPT-2 because it is free, fast, and runs easily on most machines.
generator = pipeline('text-generation', model='gpt2')

def generate_story(prompt):
    """
    This function takes a starting sentence (prompt) and lets the AI finish it.
    """
    print("AI is thinking...")
    result = generator(prompt, max_length=100, num_return_sequences=1)
    return result[0]['generated_text']
```

```python
# Let's test our tool!
my_prompt = "Once upon a time in Nairobi, a young coder discovered"
story = generate_story(my_prompt)

print("--- Generated Story ---")
print(story)
```
